## Functions

```python
# checking the skewness of the correlated data
def check_skew(df, list):
    for columns in df.select_dtypes(include=['number']).columns.to_list():
        if df[columns].skew() <= -0.5 or df[columns].skew() >= 0.5:
            list.append(columns)
            print(f"{columns} has skewness of: {df[columns].skew():.2f}")
        else:
            print(f"{columns} not skewed")


# creating a funtion to categorize the age
def categorize_age(age):
    if age < 30:
        return "20's"
    elif age < 40:
        return "30's"
    elif age < 50:
        return "40's"
    elif age < 60:
        return "50's"
    else:
        return "Above 60"

# truncating outliers
def truncate_outliers(df, variables, fold=1.5):
    for variable in variables:
        # iqr
        Q1 = df[variable].quantile(0.25)
        Q3 = df[variable].quantile(0.75)
        IQR = Q3 - Q1

        #boundry
        lower_bound = Q1 - (fold * IQR)
        upper_bound = Q3 + (fold * IQR)

        # truncate
        df[variable] = df[variable].mask(df[variable] < lower_bound, other=np.nan)
        df[variable] = df[variable].mask(df[variable] > upper_bound, other=np.nan)
        
    df= df.dropna()
    return df

# cross validation function
def cross_val(pipe, score):
    steps_dict = pipe.named_steps
    model_name = list(steps_dict.keys())[-1]
    pipe_score= cross_val_score(pipe, X_train_resampled, y_train_resampled, cv=10, scoring=score)
    print(f"{steps_dict[model_name]} CV {score}: {np.mean(pipe_score):.3f} +/- {np.std(pipe_score):.3f}")

## much better code for cross validation
# def cross_val(pipe, X, y, score):
#     try:
#         steps_dict = pipe.named_steps
#         model_name = list(steps_dict.keys())[-1]
#         pipe_score = cross_val_score(pipe, X, y, cv=10, scoring=score)
#         mean_score = np.mean(pipe_score)
#         std_score = np.std(pipe_score)
#         print(f"{steps_dict[model_name]} CV {score}: {mean_score:.3f} +/- {std_score:.3f}")
#         return mean_score, std_score
#     except Exception as e:
#         print(f"An error occurred during cross-validation: {e}")
#         return None, None

# classification report 
def performance_check(model, X, y):
    y_pred = model.predict(X)
    cm = confusion_matrix(y, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()
    print(classification_report(y, y_pred, zero_division=0))

# create a function to calculate Variance Inflation Factor
def calc_vif(X):
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)
```

# Machine Learning Smoke Classification

## Feature Splitting

```python
list_smoke= [
    'height', 'hemoglobin', 'weight', 'serum_creatinine', 'sex', 'gamma_gtp', 'waistline', 
    'sgot_alt', 'triglyceride', 'age', 'hdl_chole', 'drk_yn'
]
```

```python
X = feat_truncated[list_smoke]
y = feat_truncated['smoke_type']
```

```python
X.info()
```

```python
# checking class balance
print(y.value_counts(normalize=True))
```

```python
# splitting between train and test with stratification
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=44, stratify=y)

print('X-train', X_train.shape)
print('X-Test', X_test.shape)
```

Data is imbalance, since the size is big, will be implementing undersampling

### Undersampling

```python
# assigning a variable for under sampling
under = RandomUnderSampler(random_state=44)
```

```python
# calling the undersampling variable
X_train_resampled, y_train_resampled = under.fit_resample(X_train, y_train)
```

```python
# even distribution
print(pd.Series(y_train_resampled).value_counts(normalize=True))
```

```python
print('X-train_resampled', X_train_resampled.shape)
```

## PCA

```python
# preprocess pipe
scaler= StandardScaler()
pca_process= make_column_transformer((scaler, list_col_scale), remainder='passthrough')
X_train_scaled= pca_process.fit_transform(X_train_resampled)
```

```python
# pca
pca = PCA()
pca.fit(X_train_scaled)

# visual
plt.figure(figsize=[5, 4])
plt.plot(range(1, len(X.columns) + 1), np.cumsum(pca.explained_variance_ratio_ * 100))
plt.xlabel('Number of components')
plt.ylabel('Explained Variance Ratio - Cummulative')
plt.grid()
plt.show()
```

```python
# 95% data kept
cumsum = np.cumsum(pca.explained_variance_ratio_)
num_features = np.argmax(cumsum >= 0.95) + 1
print(f'Number of components: {num_features}')
```

## Models

```python
# preprocessing
scaler= StandardScaler()
preprocessor= make_column_transformer((scaler, list_col_scale), remainder='passthrough')
```

```python
# 13 n_components with low correlation, 5 with only moderate

logreg_pipe= make_pipeline(preprocessor, PCA(n_components=9), LogisticRegression(n_jobs=-1))

tree_pipe = make_pipeline(preprocessor, DecisionTreeClassifier())

sgd_pipe = make_pipeline(preprocessor, SGDClassifier(n_jobs=-1))

lda_pipe = make_pipeline(preprocessor, LinearDiscriminantAnalysis())
```

```python
%%time

models = [logreg_pipe, tree_pipe, sgd_pipe, lda_pipe]

for m in models:
    cross_val(m, 'accuracy')
```

```python
%%time
param= {
    'lineardiscriminantanalysis__solver': ['lsqr', 'eigen'],
    'lineardiscriminantanalysis__shrinkage': [None, 'auto', 0.5]
}

grid_lda= GridSearchCV(
    estimator=lda_pipe, param_grid=param, 
    scoring='accuracy', verbose=1
)

grid_lda.fit(X_train_resampled, y_train_resampled)
```

```python
print(f"Best parameters: {grid_lda.best_params_}")
print(f"Best cross-validation accuracy: {grid_lda.best_score_:.2f}")
```

```python
best_lda_model = grid_lda.best_estimator_
```

```python
%%time

param = {
    'sgdclassifier__loss': ['hinge', 'log_loss', 'perceptron'],
    'sgdclassifier__penalty': ['l2', 'l1', 'elasticnet'],
    'sgdclassifier__alpha': [0.0001, 0.001, 0.01],
    'sgdclassifier__learning_rate': ['constant', 'optimal', 'invscaling', 'adaptive'],
    'sgdclassifier__eta0': [0.1, 0.01, 0.001],
}
grid_sgd= GridSearchCV(
    estimator=sgd_pipe, param_grid=param, 
    scoring='accuracy', verbose=1
)

grid_sgd.fit(X_train_resampled, y_train_resampled)
```

```python
print(f"Best parameters: {grid_sgd.best_params_}")
print(f"Best cross-validation accuracy: {grid_sgd.best_score_:.3f}")
```

```python
best_sgd_model = grid_sgd.best_estimator_
```

```python

print('Train Set')
performance_check(best_lda_model, X_train, y_train)

print('Test Set')
performance_check(best_lda_model, X_test, y_test)
```

```python
print('Train Set')
performance_check(best_sgd_model, X_train, y_train)

print('Test Set')
performance_check(best_sgd_model, X_test, y_test)
```
